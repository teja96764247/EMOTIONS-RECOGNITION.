```python
import random
random.shuffle(data)
```

```python
x=[]
y=[]
for i,j in data:
  x.append(i)
  y.append(j)
```

```python
x1=np.array(x)
y1=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.2,random_state=2)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
x_train_scaled=x_train/255
x_test_scaled=x_test/255
```

```python
x_train_scaled[0]
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
num_of_classes=7
model=keras.Sequential()

model.add(keras.layers.Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(64,64,3)))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(keras.layers.Conv2D(64,kernel_size=(3,3),activation='relu'))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(keras.layers.Flatten())

model.add(keras.layers.Dense(128,activation='relu'))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(64,activation='relu'))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(num_of_classes,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train_scaled,y_train,epochs=50,validation_split=0.1)
```

```python
img_path='/content/dataset/Angry/02bc9437f479433dd373fd91f024bea7751f22504f4be12bfd31cf3d~angry.jpg'
img_pillow=Image.open(img_path)
img_color=img_pillow.convert('RGB')
img_resize=img_color.resize((64,64))
img_arr=np.array(img_resize)
img_scaled=img_arr/255
image=np.reshape(img_scaled,[1,64,64,3])
prediction=model.predict(image)
output=np.argmax(prediction)
if(output==1):
  print('AHEGAO')
elif(output==2):
  print('ANGRY')
elif(output==3):
  print('HAPPY')
elif(output==4):
  print('NEUTRAL')
elif(output==5):
  print('SAD')
elif(output==6):
  print('SURPRISE')
else:
  print('Its not possible to detect sorry!!!')
```
